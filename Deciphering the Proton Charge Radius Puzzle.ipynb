## Data Regression
Before one can analyze data using any method, it is imperative to understand some fundamentals of data science and Python. Like everything, the guide portion of this report has a disclaimer. It is assumed that the reader has a basic understanding of programming as well as algebra and statistics. In addition to general knowledge, it is assumed that the reader is familiar with Python and is capable of writing basic programs. Now for the fun.

A good place to start would be with data analysis itself. Data analysis is defined as “a process of inspecting, cleansing, transforming and modeling data to discover useful information, informing conclusions and supporting decision-making.” These tasks can be accomplished through a myriad of mathematical and computational methods, but these analyses primarily involve modeling data through one regression technique. The technique used throughout the analyses is the least-squares regression technique. A least-squares regression minimizes the sum of squared residuals (a residual being: the difference between an observed value, and the fitted value provided by a model). This method for fitting data creates a curve that is an approximation showing the correlation between two variables. For example, if one is given data sets with points and their error, it is possible to plot these points with their error bars and determine a curve of best fit. 

So now it is known that one takes some data and as if by magic, gets a curve of best fit with fit parameters and normalization parameters. What does all of this mean? Fit parameters can be thought of as variables, and, in this case, they are called the C1 and C2 (also a C3 for some analyses) parameters. These parameters refer to the coefficients of the x-terms in a function. For example, _y=mx+b_, one of the most well-known equations in math, is the equation of a line. The slope, or 'm', is the C1 parameter, which is for the $x^1$ term. This term is especially important in these analyses because the C1 parameter is used to calculate the proton charge radius. In its simplest form, the equation for the proton charge radius is $(-6*C1)^{1/2}$. The C2 parameter is the coefficient for the $x^2$ term and is the primary element of contention between different models for the data. Parameters can be called either fixed or free, which have very different effects on a regression. A "fixed" parameter is initialized (initially set) to a specific value that stays constant throughout the regression. A "free" parameter is the opposite meaning that its final value is determined by the regression, and should be the "best" number for the function to model the data.

## Monte Carlo
The Monte Carlo simulation technique, name after the Monte Carlo Casino in Monaco, involves taking fit parameters of real data, and then generating synthetic data that is analyzed. This technique allows the user to simulate data so that the error can be analyzed via histograms of the parameters.

## Bootstrapping
Bootstrapping is a technique involves taking real data sets and conducting random sampling with replacement. This method allows the user to create an unimaginable number of bootstrap data sets from one data set. This number can be calculated by taking the number of data entries and raising it to itself ($n^n$). For example, with 20 data points one is able to create 104857600000000000000000000 different data sets.

## Generate Data
***
The first step in this process of generating and then analyzing data is to generate data that is easily understood by the user. This program generates data between two numbers, then it is analyzed using a least-squares regression. The data is then plotted against the fit function, and the curve should fit the data nicely, unless the code is wrong. This first program allows the user to understand the effects of line selection, error, and number of fit parameters on the fit function.

```python
##Import Packages
import random
import matplotlib.pyplot as plt
import numpy as np
from lmfit import Model, Parameters
from scipy.interpolate import interp1d

##Function Storage
#Float Norms Function
def floats(X_Vals,C1,C2,C3,B,norm):
    f1= norm*(C1*X_Vals+C2*X_Vals**2+C3*X_Vals**3+B)
    return np.array([f1])

##Generate Values
#Define Line
Num_Points=85
le_x=np.linspace(0,1,Num_Points)
X_Vals=le_x*random.gauss(1,0.01)+random.gauss(0,0.01)
set_c1=-0.12
set_c2=0.01
b=1
#Generate Real Y-Values
y=(b+set_c1*X_Vals+set_c2*X_Vals**2)

#Initialize Y
Y=[]

#Randomize Y-Values
for q in range(0,len(y)):
    Y=np.append(Y,y[q]*random.gauss(1,0.01)+random.gauss(0,0.01))

##Regression
#Parameters
params = Parameters()
params.add('C1',  value=-0.12,vary=1)
params.add('C2',  value=0.01,vary=1)
params.add('C3',  value=0,vary=0)
params.add('B',   value=b,vary=0)
params.add('norm', value=1,vary=1)
#Result
model=Model(floats)
result=model.fit(Y,X_Vals=X_Vals,params=params,weights=1)
print(result.fit_report())
#Save Parameter Vals
Real_Param_Data=[]
for name, par in result.params.items():
    Real_Param_Data=np.append(Real_Param_Data,round(par.value,5))

##Plot
plt.figure(dpi=100)
if Num_Points<=100:
    plt.scatter(X_Vals,Y/Real_Param_Data[4],color='orange')
else:
    plt.plot(X_Vals,Y,color='orange')
plt.plot(X_Vals,Real_Param_Data[0]*X_Vals+Real_Param_Data[1]*X_Vals**2+Real_Param_Data[2]*X_Vals**3+Real_Param_Data[3],color='blue')
plt.show()
```

## Monte Carlo
***
Then this generated data can be analyzed using Monte Carlo simulations. This code uses the fit parameters from the previous least-squares regression to generate synthetic data. Synthetic data refers to the fact that the x-values are the same as the real data, but the y-values are generated using the fit parameters and the stated error. After a Monte Carlo is run for 10,000 times (or however many iterations desired), the histograms of the parameter distributions can be observed. These histograms show the mean, median, and standard deviation of the fit parameters. 

```python
##Function Storage
#Float Norms Function
def MC_floats(X,C1,C2,C3,B):
    f1= C1*X+C2*X**2+C3*X**3+B
    return np.array([f1])

##Set True Values
MC_c1,MC_c2,MC_c3,MC_b=Real_Param_Data[0],Real_Param_Data[1],Real_Param_Data[2],Real_Param_Data[3]

##Set X
X=np.linspace(0,10,Num_Points)*random.gauss(1,0.01)+random.gauss(0,0.01)

##Loop
Loop=10000
MC_Param_Data=[]
for L in range(Loop):
    Y_Vals=[]
    for q in range(0,len(X)):
        Y_Vals=np.append(Y_Vals,(MC_c1*X[q]+MC_c2*X[q]**2+MC_c3*X[q]**3+MC_b)*random.gauss(1,0.01)+random.gauss(0,0.01))
    #Regression
    params = Parameters()
    params.add('C1',  value=1,vary=1)
    params.add('C2',  value=1,vary=1)
    params.add('C3',  value=0,vary=0)
    params.add('B',  value=b,vary=0)
    model=Model(MC_floats)
    result=model.fit(Y_Vals,X=X,params=params,weights=1)
    #print(result.fit_report())
    if Loop >5:
        for name, par in result.params.items():
            MC_Param_Data=np.append(MC_Param_Data,round(par.value,5))

#Print One to Test
print(result.fit_report())
            
##Plot
MCP_c1,MCP_c2,MCP_c3,MCP_b=[],[],[],[]
for x in range(int(len(MC_Param_Data)/4)):
    MCP_c1.append(MC_Param_Data[4*x])
    MCP_c2.append(MC_Param_Data[4*x+1])
    MCP_c3.append(MC_Param_Data[4*x+2])
    MCP_b.append(MC_Param_Data[4*x+3])

##Analyze Data
params_list=[MCP_c1,MCP_c2,MCP_c3,MCP_b]
def Analyze_Graph(Name):
    new_data=(Name)#-np.mean(Name))
    print('Mean:',round(np.mean(Name),5),'')
    print('Median:',np.median(Name))
    print('SDT:',round(np.std(Name),5),'')
    print('Data Points:',len(Name))
    print()
    #Plotting
    plt.figure(dpi=100)
    plt.hist(new_data,bins=50)
    plt.title("Synthetic Data: Parameter Histogram")
    plt.xlabel('Distance From Mean')
    plt.ylabel('Frequency')
    #plt.savefig('Monte_Carlo_Synthetic_Data_(3000-Points).png')
    plt.show()
    return round(np.mean(Name),5)
MC_Mean=[]
for x in range(0,len(params_list)):
    Mean=Analyze_Graph(params_list[x])
    MC_Mean=np.append(MC_Mean,Mean)
```

## Bootstrap
***
Bootstrapping is an effective way to randomize data, and it is especially good to compare it with Monte Carlo distributions. This program uses the data that was generated in the first program that is sampled with replacement. The bootstrap can be run for 10,000 iterations (or any value), then the parameter distributions can be analyzed with their mean, median, and standard deviation.

```python
##Function Storage
def Boot_floats(X_List,C1,C2,C3,B):
    f1= C1*X_List+C2*X_List**2+C3*X_List**3+B
    return np.array([f1])

##Loop
Loop=10000
Boot_Param_Data=[]
for L in range(Loop):
    #Initialize
    Boot_Data=[]
    X_List=[]
    #Bootstrap
    for x in range(Num_Points):
        Entry_Num=np.random.randint(0,Num_Points-1)
        Boot_Data=np.append(Boot_Data,Y[Entry_Num])
        X_List=np.append(X_List,X_Vals[Entry_Num])
    
    #Regression
    params = Parameters()
    params.add('C1',  value=1,vary=1)
    params.add('C2',  value=1,vary=1)
    params.add('C3',  value=0,vary=0)
    params.add('B',  value=b,vary=0)
    model=Model(Boot_floats)
    result=model.fit(Boot_Data,X_List=X_List,params=params,weights=1)
    if Loop >5:
        for name, par in result.params.items():
            Boot_Param_Data=np.append(Boot_Param_Data,round(par.value,5))

#Print One to Test
print(result.fit_report())
            
##Plot
BootP_c1,BootP_c2,BootP_c3,BootP_b=[],[],[],[]
for x in range(int(len(Boot_Param_Data)/4)):
    BootP_c1.append(Boot_Param_Data[4*x])
    BootP_c2.append(Boot_Param_Data[4*x+1])
    BootP_c3.append(Boot_Param_Data[4*x+2])
    BootP_b.append(Boot_Param_Data[4*x+3])

##Analyze Data
params_list=[BootP_c1,BootP_c2,BootP_c3,BootP_b]
def Analyze_Graph(Name):
    new_data=(Name)#-np.mean(Name))
    print('Mean:',round(np.mean(Name),5),'')
    print('Median:',np.median(Name))
    print('SDT:',round(np.std(Name),5),'')
    print('Data Points:',len(Name))
    print()
    #Plotting
    plt.figure(dpi=100)
    plt.hist(new_data,bins=50)
    plt.title("Synthetic Data: Parameter Histogram")
    plt.xlabel('Distance From Mean')
    plt.ylabel('Frequency')
    #plt.savefig('Monte_Carlo_Synthetic_Data_(3000-Points).png')
    plt.show()
    return round(np.mean(Name),5)
Boot_Mean=[]
for x in range(0,len(params_list)):
    Mean=Analyze_Graph(params_list[x])
    Boot_Mean=np.append(Boot_Mean,Mean)
```

## Rosenfelder Least-Squares Regression
***
The least-squares regression of the Rosenfelder Data was the first method for analysis using a quadratic to model the relationship between $Q^2$ and $G_E$. Through this regression, one can find the C1 and C2 fit parameters, as well as the normalization parameters. The C1 parameter is the proton charge radius, thus this value is the “answer” of the regression. While this is true, the real area of interest for this analysis is the C2 parameter because this parameter determines the C1 parameter. This program analyzes the six sets of data used by Rosenfelder to generate fit parameters that can be used by the next program.

```python
#Graph First Time
%matplotlib inline

##Import Packages
import matplotlib.pyplot as plt
import numpy as np
from lmfit import Model, Parameters
import random

#Float Function for Modeling/Regression
def floats(Q2,ns1,ns2,nr,nb1,nb2,nm,c1,c2):
    f1= ns1*(1+c1*S1[0:,0]+(c2*S1[0:,0]**2))
    f2= ns2*(1+c1*S2[0:,0]+(c2*S2[0:,0]**2))
    f3=  nr*(1+c1*R[0:,0] +(c2*R[0:,0]**2))
    f4= nb1*(1+c1*B1[0:,0]+(c2*B1[0:,0]**2))
    f5= nb2*(1+c1*B2[0:,0]+(c2*B2[0:,0]**2))
    f6=  nm*(1+c1*M[0:,0] +(c2*M[0:,0]**2))
    return np.concatenate((f1,f2,f3,f4,f5,f6))

##Import Data Files
S1=np.loadtxt('+S1sort.dat')
S2=np.loadtxt('+S2sort.dat')
R =np.loadtxt('+Rsort.dat' )
B1=np.loadtxt('+B1sort.dat')
B2=np.loadtxt('+B2sort.dat')
M =np.loadtxt('+Msort.dat' )

Full_Param_Data=[]
##Generate Values
#Q2
Q2=np.concatenate((S1[0:,0],S2[0:,0],R[0:,0],B1[0:,0],B2[0:,0],M[0:,0]))
#GE
GE=np.concatenate((S1[0:,1],S2[0:,1],R[0:,1],B1[0:,1],B2[0:,1],M[0:,1]))
#Error
Error=np.concatenate((S1[0:,2],S2[0:,2],R[0:,2],B1[0:,2],B2[0:,2],M[0:,2]))
#Coulomb Correction
coulomb_corr=np.concatenate((S1[0:,3],S2[0:,3],R[0:,3],B1[0:,3],B2[0:,3],M[0:,3]))

#Use CC (1 or 0)
Use_CC=0
if Use_CC==1:
    GE=GE-coulomb_corr
                          
##Regression
#Parameters
params = Parameters()
params.add('c1',  value=-0.12,vary=1)
params.add('c2',  value=0.014,vary=1)
params.add('ns1', value=1,vary=1)
params.add('ns2', value=1,vary=1)
params.add('nr',  value=1,vary=1)
params.add('nb1', value=1,vary=1)
params.add('nb2', value=1,vary=1)
params.add('nm',  value=1,vary=1)                     
#Result
model=Model(floats)
result=model.fit(GE,Q2=Q2,params=params,weights=1/Error)
#Save Params to List
Real_Params=[]
for name, par in result.params.items():
    Param=round(par.value,5)
    Real_Params.append(Param)
#Fit Report 
print('Radius:',round((((-6*Real_Params[0])**2)**0.25),5))
print(result.fit_report())

##Plot
#Formatting
plt.figure(dpi=100)
#plt.title('$Q^2$ vs $G_E$: Function vs Data Points',fontname='Times New Roman',fontsize=15)
plt.xlabel('$Q^2$ [fm$^{-2}$]',fontname='Times New Roman',fontsize=14)
plt.ylabel('$G_E$',fontname='Times New Roman',fontsize=14)
#Scatter Plot with Error Bars
#With CC
if Use_CC==1:
    plt.errorbar(S1[:,0],(S1[:,1]-S1[:,3])/Real_Params[2],yerr=S1[:,2],fmt='o',label='S1')
    plt.errorbar(S2[:,0],(S2[:,1]-S2[:,3])/Real_Params[3],yerr=S2[:,2],fmt='o',label='S2')
    plt.errorbar(R[:,0],(R[:,1]-R[:,3])/Real_Params[4],yerr=R[:,2],fmt='o',label='R')
    plt.errorbar(B1[:,0],(B1[:,1]-B1[:,3])/Real_Params[5],yerr=B1[:,2],fmt='o',label='B1')
    plt.errorbar(B2[:,0],(B2[:,1]-B2[:,3])/Real_Params[6],yerr=B2[:,2],fmt='o',label='B2')
    plt.errorbar(M[:,0],(M[:,1]-M[:,3])/Real_Params[7],yerr=M[:,2],fmt='o',label='M')
#Without CC
else:
    plt.errorbar(S1[:,0],S1[:,1]/Real_Params[2],yerr=S1[:,2],fmt='o',label='S1')
    plt.errorbar(S2[:,0],S2[:,1]/Real_Params[3],yerr=S2[:,2],fmt='o',label='S2')
    plt.errorbar(R[:,0],R[:,1]/Real_Params[4],yerr=R[:,2],fmt='o',label='R')
    plt.errorbar(B1[:,0],B1[:,1]/Real_Params[5],yerr=B1[:,2],fmt='o',label='B1')
    plt.errorbar(B2[:,0],B2[:,1]/Real_Params[6],yerr=B2[:,2],fmt='o',label='B2')
    plt.errorbar(M[:,0],M[:,1]/Real_Params[7],yerr=M[:,2],fmt='o',label='M')

plt.xlim(0,1.05)
Q2=np.linspace(0,1.05)

#Plot Fit Function
#plt.plot(Q2,1+Real_Params[0]*Q2+Real_Params[1]*Q2**2,color='red')

#Plot Standard Fits
#plt.plot(Q2,1+-0.12873976*Q2+0.014*Q2**2,color='blue',label='With CC')
#plt.plot(Q2,1+-0.12748012*Q2+0.014*Q2**2,color='orange',label='Without CC')

plt.plot(Q2,1+-0.11984520*Q2+0.00536785*Q2**2,color='blue',label='With CC')
plt.plot(Q2,1+-0.11484108*Q2+0.00173400*Q2**2,color='orange',label='Without CC')

plt.legend()
plt.show()
```

## Rosenfelder Monte Carlo
***
This program generates synthetic data sets based on the quadratic from a least squares regression. The real data comes from Rosenfelder's compiled data sets, but is copied, the $G_E$ is replaced by the function value evaluated at $Q^2$. Each point is then randomized based on the synthetic data and the reported error via numpy's gauss function. Then a regression of the data is done and the parameters of the regression are appended list. The parameter list is then analyzed by the next block of code that generates histograms of the parameters.

```python
##Graph First Time
%matplotlib inline

#Monte Carlo Float Norms Function
def MC_floats(MC_Q2,ns1,ns2,nr,nb1,nb2,nm,c1,c2):
    f1= ns1*(1+c1*S1[0:,0]+(c2*S1[0:,0]**2))
    f2= ns2*(1+c1*S2[0:,0]+(c2*S2[0:,0]**2))
    f3=  nr*(1+c1*R[0:,0] +(c2*R[0:,0]**2))
    f4= nb1*(1+c1*B1[0:,0]+(c2*B1[0:,0]**2))
    f5= nb2*(1+c1*B2[0:,0]+(c2*B2[0:,0]**2))
    f6=  nm*(1+c1*M[0:,0] +(c2*M[0:,0]**2))
    return np.concatenate((f1,f2,f3,f4,f5,f6))

##Set True Values
MC_c1,MC_c2=Real_Params[0],Real_Params[1]
print('C1 Parameter:', MC_c1)
print('C2 Parameter:', MC_c2)

#Set Randomization
Set_Rand=0.005

##Loop
Loop=10000
MC_Param_Data=[]
for L in range(Loop):
    ##Get Q2
    MC_Q2=np.concatenate((S1[0:,0]*random.gauss(1,Set_Rand),S2[0:,0]*random.gauss(1,Set_Rand),R[0:,0]*random.gauss(1,Set_Rand),B1[0:,0]*random.gauss(1,Set_Rand),B2[0:,0]*random.gauss(1,Set_Rand),M[0:,0]*random.gauss(1,Set_Rand)))
    #Get GE
    MC_GE=[]
    for q in range(0,len(MC_Q2)):
        MC_GE=np.append(MC_GE,random.gauss((1+MC_c1*MC_Q2[q]+MC_c2*MC_Q2[q]**2),Error[q]))
    #Regression
    params = Parameters()
    params.add('c1',  value=-0.12,vary=1)
    params.add('c2',  value=random.gauss(0.014,0.00),vary=1)
    params.add('ns1', value=1,vary=1)
    params.add('ns2', value=1,vary=1)
    params.add('nr',  value=1,vary=1)
    params.add('nb1', value=1,vary=1)
    params.add('nb2', value=1,vary=1)
    params.add('nm',  value=1,vary=1)
    model=Model(MC_floats)
    result=model.fit(MC_GE,MC_Q2=MC_Q2,params=params,weights=1)
    #Save to List
    for name, par in result.params.items():
        MC_Param_Data=np.append(MC_Param_Data,round(par.value,5))
    MC_Param_Data=np.append(MC_Param_Data,result.chisqr)

#Print One to Test (Testing Purposes)
print('Radius:',round((((-6*Real_Params[-8])**2)**0.25),5))
print(result.fit_report())

##Plot
MCP_c1,MCP_c2,MCP_ns1,MCP_ns2,MCP_nr,MCP_nb1,MCP_nb2,MCP_nm,MCP_chisqr=[],[],[],[],[],[],[],[],[]
for x in range(Loop):
    MCP_c1.append(((-6*MC_Param_Data[9*x])**2)**0.25)
    MCP_c2.append(MC_Param_Data[9*x+1])
    MCP_ns1.append(MC_Param_Data[9*x+2])
    MCP_ns2.append(MC_Param_Data[9*x+3])
    MCP_nr.append(MC_Param_Data[9*x+4])
    MCP_nb1.append(MC_Param_Data[9*x+5])
    MCP_nb2.append(MC_Param_Data[9*x+6])
    MCP_nm.append(MC_Param_Data[9*x+7])
    MCP_chisqr.append(MC_Param_Data[9*x+8])

##Analyze Data
MC_params_list=[MCP_c1,MCP_c2,MCP_chisqr,MCP_ns1,MCP_ns2,MCP_nr,MCP_nb1,MCP_nb2,MCP_nm]
x_axis=['Proton Charge Radius [fm]','C2 Parameter','Chi Square','S1 Norm','S2 Norm','R Norm','B1 Norm','B2 Norm','M Norm']
def Analyze_Graph(Name,MC_Q2_Axis):
    new_data=(Name)#-np.mean(Name))
    print('Mean:',round(np.mean(Name),5),'fm')
    print('Median:',round(np.median(Name),5),'fm')
    print('\u03C3:',round(np.std(Name),5),'fm')
    print('Iterations:',len(Name))
    #Plotting
    plt.figure(dpi=100)
    plt.hist(new_data,bins=60,color='blue')
    plt.ylabel('Frequency')
    plt.xlabel(MC_Q2_Axis)
    #plt.savefig('Rosen_Monte.Carlo.png')
    plt.show()
    return round(np.mean(Name),5)
MC_Mean=[]
for x in range(0,len(MC_params_list)):
    Mean=Analyze_Graph(MC_params_list[x],x_axis[x])
    MC_Mean=np.append(MC_Mean,Mean)
```

## Rosenfelder Bootstrap 

This program randomly samples actual data and does a least-squares regression of each set of 85 points for the number of loops. The real data comes from the six sets of data that are used by Rosenfelder. The program imports the data from the six data files, which is then copied by numpy. Then the number of loops is set and the main loop begins with Data, compiled list of selected data entries, and the lists of $Q^2$ values. Then a random integer is selected and the corresponding $Q^2$ value and full data entry are saved to a list. This repeats for all six sets. Then the $Q^2$, $G_E$, and error values are separated and put into numpy arrays. The regression uses these three lists do determine the parameters, which is when the lists of $Q^2$ values are used. Each list corresponds to a data set, thus each list is used as the input for the function generated by the regression; this generates the normalization parameters. The parameter values are appended to a list that is used by the second block of code to generate histograms showing the distribution of the parameters.

```python
##### Graph First Time
%matplotlib inline

##Import Packages
import random
import matplotlib.pyplot as plt
import numpy as np
from lmfit import Model, Parameters

#Float Function for Modeling/Regression
def float_norms(q2,c1,c2,c3,Set_1,Set_2,Set_3):
    f1= Set_1*(1+c1*(Set1[0:,0]*25.7)+c2*(Set1[0:,0]*25.7)**2+c3*(Set1[0:,0]*25.7)**3)
    f2= Set_2*(1+c1*(Set2[0:,0]*25.7)+c2*(Set2[0:,0]*25.7)**2+c3*(Set2[0:,0]*25.7)**3)
    f3= Set_3*(1+c1*(Set3[0:,0]*25.7)+c2*(Set3[0:,0]*25.7)**2+c3*(Set3[0:,0]*25.7)**3)
    return np.concatenate((f1,f2,f3))

##Import Data (Data Used by Rosenfelder)
Set1=np.loadtxt('New_Data(0.495_GeV).dat')
Set2=np.loadtxt('New_Data(0.330_GeV).dat')
Set3=np.loadtxt('New_Data(0.195_GeV).dat')

#Initialize Arrays
Params=[]
#Paramaters
params = Parameters()
params.add('c1',  value=-0.12,vary=1)
#Fixed Parameters (MAMI)
#params.add('c2',  value=2.59/120,vary=0)
#params.add('c3',  value=-29.8/5040,vary=0)
#Rosenfelder Parameters
params.add('c2', value=random.gauss(0.014,0.00),vary=0)
params.add('c3', value=0,vary=0)
#Normailization Parameters
params.add('Set_1', value=1,vary=1)
params.add('Set_2', value=1,vary=1)
params.add('Set_3', value=1,vary=1)

#Concatenation
q2   = np.concatenate((Set1[0:,0],Set2[0:,0],Set3[0:,0]))
data = np.concatenate((Set1[0:,3],Set2[0:,3],Set3[0:,3]))
err  = np.concatenate(((Set1[0:,2]**2+Set1[0:,5]**2)**0.5,(Set2[0:,2]**2+Set2[0:,5]**2)**0.5,(Set3[0:,2]**2+Set3[0:,5]**2)**0.5))
#err  = np.concatenate((Set1[0:,2],Set2[0:,2],Set3[0:,2]))
q2=q2*25.7

#Create Model
model=Model(float_norms)
result=model.fit(data,q2=q2,params=params,weights=1/err)
#Model Report
for name, par in result.params.items():
    Param=round(par.value,5)
    Params.append(Param)
print('Radius:',round((((-6*Params[0])**2)**0.25),5))
print(result.fit_report())

#Plot
plt.figure(dpi=100)
plt.title('$Q^2$ vs $G_E$: Function vs Data Points',fontname='Times New Roman',fontsize=15)
plt.xlabel('$Q^2$ [fm$^{-2}$]',fontname='Times New Roman')
plt.ylabel('$G_E$',fontname='Times New Roman')
plt.errorbar(Set1[:,0]*25.7,Set1[:,3]/Params[3],yerr=Set1[:,2],fmt='o',label='Set1')
plt.errorbar(Set2[:,0]*25.7,Set2[:,3]/Params[4],yerr=Set2[:,2],fmt='o',label='Set2')
plt.errorbar(Set3[:,0]*25.7,Set3[:,3]/Params[5],yerr=Set3[:,2],fmt='o',label='Set3')
plt.xlim(0,0.45)
q2.sort()
plt.plot(q2,1+Params[0]*q2+Params[1]*q2**2+Params[2]*q2**3,color='blue')
q2=np.linspace(0,0.44)
plt.plot(q2,1+Params[0]*q2+Params[1]*q2**2+Params[2]*q2**3,color='red')
plt.legend()
plt.show()
```

```python
#Graph First Time
%matplotlib inline

#Float Norms
def float_norms(Boot_Q2,c1,c2,c3,Set_1,Set_2,Set_3,ns1,ns2,nr,nb1,nb2,nm):
    fSet1= Set_1*(1+c1*Set1_List+(c2*(Set1_List)**2+c3*(Set1_List)**3))
    fSet2= Set_2*(1+c1*Set2_List+(c2*(Set2_List)**2+c3*(Set2_List)**3))
    fSet3= Set_3*(1+c1*Set3_List+(c2*(Set3_List)**2+c3*(Set3_List)**3))
    f1= ns1*(1+c1*S1_List+(c2*S1_List**2)+(c3*S1_List**3))
    f2= ns2*(1+c1*S2_List+(c2*S2_List**2)+(c3*S2_List**3))
    f3=  nr*(1+c1*R_List +(c2*R_List**2) +(c3*R_List**3))
    f4= nb1*(1+c1*B1_List+(c2*B1_List**2)+(c3*B1_List**3))
    f5= nb2*(1+c1*B2_List+(c2*B2_List**2)+(c3*B2_List**3))
    f6=  nm*(1+c1*M_List +(c2*M_List**2) +(c3*M_List**3))
    return np.concatenate((fSet1,fSet2,fSet3,f1,f2,f3,f4,f5,f6))

q2   = np.concatenate((Set1[0:,0]*25.7,Set2[0:,0]*25.7,Set3[0:,0]*25.7,S1[0:,0],S2[0:,0],R[0:,0],B1[0:,0],B2[0:,0],M[0:,0]))
data = np.concatenate((Set1[0:,3],Set2[0:,3],Set3[0:,3],S1[0:,1],S2[0:,1],R[0:,1],B1[0:,1],B2[0:,1],M[0:,1]))
err  = np.concatenate((Set1[0:,2],Set2[0:,2],Set3[0:,2],S1[0:,2],S2[0:,2],R[0:,2],B1[0:,2],B2[0:,2],M[0:,2]))

##Loop
Loop=1000
#Initialize Out Loop
Params,X=[],0
while X<(Loop):
    #Initialize In Loop
    Q2,GE,Error,C_Corr,Set_Num=[],[],[],[],[]
    #Bootstrap
    for Z in range(len(q2)):
        Rand_Ele=np.random.randint(0,len(q2))
        Q2=np.append(Q2,q2[Rand_Ele])
        GE=np.append(GE,data[Rand_Ele])
        Error=np.append(Error,err[Rand_Ele])
        Set_Num.append(Num[Rand_Ele])
    #Sort Data (By Set)
    Set1_List,Set2_List,Set3_List,S1_List,S2_List,R_List,B1_List,B2_List,M_List=[],[],[],[],[],[],[],[],[]
    Set1_Boot_GE,Set2_Boot_GE,Set3_Boot_GE,S1_Boot_GE,S2_Boot_GE,R_Boot_GE,B1_Boot_GE,B2_Boot_GE,M_Boot_GE=[],[],[],[],[],[],[],[],[]
    Set1_Boot_Error,Set2_Boot_Error,Set3_Boot_Error,S1_Boot_Error,S2_Boot_Error,R_Boot_Error,B1_Boot_Error,B2_Boot_Error,M_Boot_Error=[],[],[],[],[],[],[],[],[]
    for B in range(len(Q2)):
        if Set_Num[B]==11:
            Set1_List=np.append(Set1_List,Q2[B])
            Set1_Boot_GE=np.append(Set1_Boot_GE,GE[B])
            Set1_Boot_Error=np.append(Set1_Boot_Error,Error[B])
        if Set_Num[B]==22:
            Set2_List=np.append(Set2_List,Q2[B])
            Set2_Boot_GE=np.append(Set2_Boot_GE,GE[B])
            Set2_Boot_Error=np.append(Set2_Boot_Error,Error[B])
        if Set_Num[B]==33:
            Set3_List=np.append(Set3_List,Q2[B])
            Set3_Boot_GE=np.append(Set3_Boot_GE,GE[B])
            Set3_Boot_Error=np.append(Set3_Boot_Error,Error[B])
        if Set_Num[B]==1:
            S1_List=np.append(S1_List,Q2[B])
            S1_Boot_GE=np.append(S1_Boot_GE,GE[B])
            S1_Boot_Error=np.append(S1_Boot_Error,Error[B])
        if Set_Num[B]==2:
            S2_List=np.append(S2_List,Q2[B])
            S2_Boot_GE=np.append(S2_Boot_GE,GE[B])
            S2_Boot_Error=np.append(S2_Boot_Error,Error[B])
        if Set_Num[B]==3:
            R_List=np.append(R_List,Q2[B])
            R_Boot_GE=np.append(R_Boot_GE,GE[B])
            R_Boot_Error=np.append(R_Boot_Error,Error[B])
        if Set_Num[B]==4:
            B1_List=np.append(B1_List,Q2[B])
            B1_Boot_GE=np.append(B1_Boot_GE,GE[B])
            B1_Boot_Error=np.append(B1_Boot_Error,Error[B])
        if Set_Num[B]==5:
            B2_List=np.append(B2_List,Q2[B])
            B2_Boot_GE=np.append(B2_Boot_GE,GE[B])
            B2_Boot_Error=np.append(B2_Boot_Error,Error[B])
        if Set_Num[B]==6:
            M_List=np.append(M_List,Q2[B])
            M_Boot_GE=np.append(M_Boot_GE,GE[B])
            M_Boot_Error=np.append(M_Boot_Error,Error[B])
    Boot_Q2=np.concatenate((Set1_List,Set2_List,Set3_List,S1_List,S2_List,R_List,B1_List,B2_List,M_List))
    Boot_GE=np.concatenate((Set1_Boot_GE,Set2_Boot_GE,Set3_Boot_GE,S1_Boot_GE,S2_Boot_GE,R_Boot_GE,B1_Boot_GE,B2_Boot_GE,M_Boot_GE))
    Boot_Error=np.concatenate((Set1_Boot_Error,Set2_Boot_Error,Set3_Boot_Error,S1_Boot_Error,S2_Boot_Error,R_Boot_Error,B1_Boot_Error,B2_Boot_Error,M_Boot_Error))
    #Check to Make Sure Each Set Does Not Have Zero Entries
    if len(Set1_List)==0 or len(Set2_List)==0 or len(Set3_List)==0 or len(S1_List)==0 or len(S2_List)==0 or len(R_List)==0 or len(B1_List)==0 or len(B2_List)==0 or len(M_List)==0:
        continue

    #Paramaters
    params = Parameters()
    params.add('c1',  value=-0.12,vary=1)
    #ISR Parameters
    #params.add('c2',  value=random.gauss(2.59/120,0.194/120),vary=0)
    #params.add('c3',  value=random.gauss(-29.8/5040,14.71/5040),vary=0)
    #Rosenfelder Parameters
    params.add('c2', value=random.gauss(0.014,0.004),vary=1)
    params.add('c3', value=0,vary=0)
    #Normailization Parameters
    params.add('Set_1', value=1,vary=1)
    params.add('Set_2', value=1,vary=1)
    params.add('Set_3', value=1,vary=1)
    params.add('ns1', value=1,vary=1)
    params.add('ns2', value=1,vary=1)
    params.add('nr',  value=1,vary=1)
    params.add('nb1', value=1,vary=1)
    params.add('nb2', value=1,vary=1)
    params.add('nm',  value=1,vary=1)
    #Create Model
    model=Model(float_norms)
    result=model.fit(Boot_GE,Boot_Q2=Boot_Q2,params=params,weights=1/Boot_Error)
    #Save Params to List
    Mini_List=[]
    for name, par in result.params.items():
        Param=round(par.value,5)
        Mini_List.append(Param)
    Mini_List.append(result.chisqr)
    Params.append(Mini_List)
    X+=1

#Print to Test
#print('Radius:',round((((-6*Mini_List[0])**2)**0.25),5))
#print(result.fit_report())

##Plot 
#Separate Data
c1,c2,ns1,ns2,nr,nb1,nb2,nm,chisqr=[],[],[],[],[],[],[],[],[]
for x in range(Loop):
    c1=np.append(c1,Params[x][0])
    c2=np.append(c2,Params[x][1])
    c3=np.append(ns1,Params[x][2])
    ns1=np.append(ns2,Params[x][3])
    chisqr=np.append(chisqr,Params[x][4])
#Plotting
c1=(((-6*c1)**2)**0.25)
params_list=[c1,c2,c3,chisqr,ns1,ns2,nr,nb1,nb2,nm]
x_axis=['Proton Charge Radius [fm]','C2 Parameter','C3 Parameter','Chi Square','S1 Norm','S2 Norm','R Norm','B1 Norm','B2 Norm','M Norm']
def Analyze_Graph(Name,X_Axis):
    new_data=(Name)#-np.mean(Name))
    print('Mean:',round(np.mean(Name),4),'fm')
    print('Median:',round(np.median(Name),4),'fm')
    print('\u03C3:',round(np.std(Name),4),'fm')
    print('Iterations:',len(Name))
    print('Fit Parameters: 4')
    #Plotting
    plt.figure(dpi=100)
    plt.hist(new_data,bins=60,color='blue')
    plt.ylabel('Frequency',fontname='Times New Roman',fontsize=15)
    plt.xlabel(X_Axis,fontname='Times New Roman',fontsize=15)
    #plt.savefig('Rosen_Bootstrap(10000-Points).png')
    plt.show()
for x in range(0,2):#len(params_list)):
    Analyze_Graph(params_list[x],x_axis[x])
```

# Conclusion
***
All of these analyses show that the C2 parameter has a direct correlation with the calculated PCR, thus how the C2 parameter is selected directly effects what the PCR will be. This is where bias comes into play. One can do a regression of electron-proton scattering data and get a PCR value that is lower than expected. So one can minorly adjust the C2 parameter and get a radius that is larger. This circular logic could be a contributing factor to the proton charge radius puzzle. While the proton charge radius puzzle still persists,CODATA, an international organization that agrees upon physical constants, has changed the stated value of proton charge radius. This change from 0.8751 fm to 0.8414 fm resembles the muonic hydrogen result than the many scattering results.
